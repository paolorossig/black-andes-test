```python
from src.slgame import SnakesLaddersGame
```

```python
ladders = {3: 16, 5: 7, 15: 25, 18: 20, 21: 32}
snakes = {12: 2, 14: 11, 17: 4, 31: 19, 35: 22}

game = SnakesLaddersGame(36, snakes, ladders)
```

```python
# First Simulation
simulation = game.run_2p_simulation(10000)
print("Player 1 wins with a probability of", simulation["prob_p1_win"] * 100, "%")
print("On average, ", simulation["avg_ladders_landed"], "snakes were landed.")
print("On average, It takes", simulation["avg_rolls"], "rolls to complete a game.")
```

```python
# Second Simulation
simu_2 = game.run_2p_simulation(iterations=10000, prob_take_ladder=0.5)
print("On average, It takes", simu_2["avg_rolls"], "rolls to complete a game.")
```
