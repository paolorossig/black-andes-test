# Simulation problem

This problem is about an old table game: Snakes & Ladders.
A simulation of 2 players was performed to answer the questions.

```python
from src.slgame import SnakesLaddersGame
```

```python
ladders = {3: 16, 5: 7, 15: 25, 18: 20, 21: 32}
snakes = {12: 2, 14: 11, 17: 4, 31: 19, 35: 22}

game = SnakesLaddersGame(36, snakes, ladders)
```

1. In a two-person game, what is the probability that the player who starts the game wins?
2. On average, how many snakes are landed on in each game?

```python
# First Simulation
simulation = game.run_2p_simulation(10000)
print("Player 1 wins with a probability of {:.2f} %".format(simulation["prob_p1_win"] * 100))
print("On average, {:.1f} snakes were landed.".format(simulation["avg_snakes_landed"]))
print("On average, It takes {:.1f} rolls to complete a game.".format(simulation["avg_rolls"]))
```

3. If each time a player landed on a ladder and there was only a 50% chance they could take it, what is the average number of rolls needed to complete a game?

```python
# Second Simulation
simu_2 = game.run_2p_simulation(iterations=10000, prob_take_ladder=0.5)
print("On average, It takes {:.1f} rolls to complete a game.".format(simu_2["avg_rolls"]))
```

5. In a different attempt to change the odds of the game, instead of starting Player 2 on a different square, you decide to give Player 2 immunity to the first snake that they land on. What is the approximate probability that Player 1 wins now?
